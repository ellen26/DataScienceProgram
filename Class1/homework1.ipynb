```python
from numpy import percentile
quartiles = percentile(data['Population'], [25, 50, 75])
data_min, data_max = data['Population'].min(), data['Population'].max()
# print 5-number summary
print('Min: ', data_min)
print('Q1:',  quartiles[0])
print('Median:' , quartiles[1])
print('Q3:' , quartiles[2])
print('Max:' , data_max)
print('SD:', data['Population'].std())
```

```python
quartiles = percentile(data['Murder.Rate'], [25, 50, 75])
data_min, data_max = data['Murder.Rate'].min(), data['Murder.Rate'].max()
# print 5-number summary
print('Min: ', data_min)
print('Q1:',  quartiles[0])
print('Median:' , quartiles[1])
print('Q3:' , quartiles[2])
print('Max:' , data_max)
print('SD:', data['Murder.Rate'].std())
```
